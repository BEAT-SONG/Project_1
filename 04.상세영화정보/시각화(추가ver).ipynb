```python
import pandas as pd

mv = pd.read_csv('./movie_info.csv')
mv_list = pd.read_csv('./total_movie_list_tmp.csv')
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
import platform

plt.rc('font', family = 'Malgun Gothic')
```

```python
mv_list = mv_list.rename(columns={'대표코드':'영화코드'})
mv_list
```

# - 그룹화 합계

```python
list_name_sum = mv_list.groupby(['영화명(국문)']).sum()
list_name_sum
```

```python
# list_name_sum2 = mv_list.groupby('영화명(국문)')['관객수'].sum()
# list_name_sum2
```

```python
list_name_rank = list_name_sum.sort_values(by=['관객수'], ascending=False)
list_name_rank
```

# - merge

```python
mv['영화코드'] = mv['영화코드'].apply(str)
```

```python
merge_left = pd.merge(list_name_rank, mv, on= '영화명(국문)', how = 'left')
merge_left
```

# * 장르

```python
merge_genres = merge_left.groupby(['장르'], as_index=False).count().sort_values(by=['영화명(국문)'], ascending=False)
merge_genres 
```

```python
asd = merge_left.groupby(['영화명(국문)','장르'], as_index = False).sum()
```

```python
ss = (asd['장르'] == '드라마')|(asd['장르'] == '애니메이션')|(asd['장르'] == '액션')|(asd['장르'] == '코미디')| (asd['장르'] == '멜로/로맨스')
ss = asd[ss]
ss
```

```python
en = ss[['장르','관객수']]
en
```

```python
merge_genres = merge_genres.rename(columns={'영화명(국문)':'빈도수'})
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x="장르", y="빈도수", data=merge_genres)
plt.title("박스오피스 장르 빈도수 분석")
plt.show()
```

## 배우

```python
merge_left_2
```

```python
actor_list = []
```

```python
for i in merge_left_2['배우3']:
    actor_list.append(i)
```

```python
len(actor_list)
```

```python
actors = pd.DataFrame(actor_list,
             columns = ['배우'])
actors.groupby(['배우'], as_index=False).count().sort_values(by=['배우'], ascending=False)
```

```python
actors['count'] = actors.groupby(['배우']).배우.transform('count')
```

```python
actors_cnt = actors.groupby(['배우'], as_index=False).count().sort_values(by=['count'], ascending=False)
actors_cnt
```

```python
actors_cnt = actors_cnt.drop(actors_cnt.index[0])
actors_cnt_rank = actors_cnt.head(20)
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x="배우", y="count", data=actors_cnt_rank)
plt.title("박스오피스 배우 빈도수 분석")
plt.show()
```
