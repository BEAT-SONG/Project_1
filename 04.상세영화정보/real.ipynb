```python
yourkey='1f84ed9279086ccfa21c6ef005b89a5f'

t_dict = {'영화코드': [], '영화명(국문)': [], '상영시간': [], '장르': [], '감독': [], '배우1':[], '배우2':[], '배우3':[], '참여 영화사1':[], '참여 영화사2':[] }


for code in codes_4000 :

    code_str = str(code)
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key='+yourkey+'&movieCd='+code_str
    res = requests.get(url)
    text= res.text

    d = json.loads(text)


    x = d['movieInfoResult']['movieInfo']

    info_list1 = []
    info_list2 = []
    info_list3 = []
    info_list4 = []
    info_list5 = []
    info_list6 = []
    info_list7 = []


    temp1 = x['movieCd']
    info_list1.append(temp1)
    info_list1
    while len(info_list1) < 1 :
        info_list1.append('없음')
    a = info_list1

    temp2 = x['movieNm']
    info_list2.append(temp2)
    info_list2
    while len(info_list2) < 1 :
        info_list2.append('없음')
    b = info_list2

    temp3 = x['showTm']
    info_list3.append(temp3)
    info_list3
    while len(info_list3) < 1 :
        info_list3.append('없음')
    c = info_list3

    temp4 = x['directors']
    for director in temp4 :
        info_list4.append(director['peopleNm'])
    while len(info_list4) < 1 :
        info_list4.append('없음')
    e = info_list4[:1]

    temp5 = x['genres']
    for genre in temp5 :
        info_list5.append(genre['genreNm'])
    while len(info_list5) < 1 :
        info_list5.append('없음')
    f = info_list5[:1]

    temp6 = x['actors']
    for actor in temp6 :
        info_list6.append(actor['peopleNm'])
    while len(info_list6) < 3 :
        info_list6.append('없음')
    g = info_list6[:3]

    temp7 = x['companys']
    for company in temp7 :
        info_list7.append(company['companyNm'])
    while len(info_list7) < 2 :
        info_list7.append('없음')
    h = info_list7[:2]


    t_dict['영화코드'].append(a[0])
    t_dict['영화명(국문)'].append(b[0])
    t_dict['상영시간'].append(c[0])
    t_dict['감독'].append(e[0])
    t_dict['장르'].append(f[0])
    t_dict['배우1'].append(g[0])
    t_dict['배우2'].append(g[1])
    t_dict['배우3'].append(g[2])
    t_dict['참여 영화사1'].append(h[0])
    t_dict['참여 영화사2'].append(h[1])
    
data = pd.DataFrame(t_dict)

data.to_csv("./info_collect/movie_info_4000.csv", mode='w', encoding='utf-8', index=False) 
```
