```python
from datetime import datetime, timedelta
import requests
import json
import pandas as pd

#날짜를 문자열로 변환하는 함수
def date_string(date):
    date_list = str(date).split()[0].split('-')
    date_str = date_list[0]+date_list[1]+date_list[2]
    return date_str

yourkey='db34324ef66ec08e25d5fb7ca93bea96' #((키 넣어주시면 됩니다!))

#원하는 날짜 +1 해주세요. 2018년 전체를 하고싶다면 (2019,1,1) ((날짜 수정해주시면 됩니다!))
date = datetime(2004,1,3)
date_str = date_string(date)
print(date_str)

# 2018년 1월1일까지만 할것이다 ((날짜 수정해주시면 됩니다!))
while date_str != '20040101':
    date = date - timedelta(days=1)
    date_str = date_string(date)
    # print(date_str)

    url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key='+yourkey+'&targetDt='+date_str
    res = requests.get(url)
    text= res.text

    d = json.loads(text)

    df = pd.DataFrame(columns=['순위',
                               '순위 증감분', 
                               '신규진입여부',
                               '대표코드',
                               '영화명(국문)',
                               '개봉일',
                               '매출액',
                               '매출총액 대비 매출비율',
                               '전일 대비 매출액 증감분',
                               '전일 대비 매출액 증감비율',
                               '누적매출액',
                               '관객수',
                               '전일 대비 관객수 증감분',
                               '전일 대비 관객수 증감 비율',
                               '누적관객수',
                               '상영 스크린 수',
                               '상영 횟수'])

    x=d['boxOfficeResult']['showRange'] #박스오피스 일자 추출
    

    for b in d['boxOfficeResult']['dailyBoxOfficeList']:
        df.append(pd.DataFrame([[x,b['rank'],
                       b['rankInten'],
                       b['rankOldAndNew'],
                       b['movieCd'],
                       b['movieNm'],
                       b['openDt'],
                       b['salesAmt'],
                       b['salesShare'],
                       b['salesInten'],
                       b['salesChange'],
                       b['salesAcc'],
                       b['audiCnt'],
                       b['audiInten'],
                       b['audiChange'],
                       b['audiAcc'],
                       b['scrnCnt'],
                       b['showCnt']]], columns=['일자','순위',
                               '순위 증감분', 
                               '신규진입여부',
                               '대표코드',
                               '영화명(국문)',
                               '개봉일',
                               '매출액',
                               '매출총액 대비 매출비율',
                               '전일 대비 매출액 증감분',
                               '전일 대비 매출액 증감비율',
                               '누적매출액',
                               '관객수',
                               '전일 대비 관객수 증감분',
                               '전일 대비 관객수 증감 비율',
                               '누적관객수',
                               '상영 스크린 수',
                               '상영 횟수']), ignore_index=True )
    
    print(df)
        
    # rnum은 순번을 출력하는 것이기 때문에 안넣었습니다

    # 파일 이름 변경하는 곳 ((파일이름 수정해주시면 됩니다!))
    
    # data.to_csv("2001_movie_day.csv", mode='a', encoding='utf-8', index=False)
```
