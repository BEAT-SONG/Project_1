```python
import pandas as pd

mv = pd.read_csv('./movie_info.csv')
mv_list = pd.read_csv('./total_movie_list_tmp.csv')
```

```python
mv_list = mv_list.rename(columns={'대표코드':'영화코드'})
mv_list
```

```python
mv_list = mv_list.drop(['Unnamed: 0'], axis=1)
mv_list
```

```python
mv
```

```python
list_name_rank = mv_list.groupby(['영화명(국문)']).sum().sort_values(by=['관객수'], ascending=False)
list_name_rank
```

# - merge

```python
mv['영화코드'] = mv['영화코드'].apply(str)
```

```python
merge_left_name = pd.merge(mv_list, mv, on= '영화명(국문)', how = 'left')
merge_left_name
```

```python
merge_left_code = pd.merge(mv_list, mv, on= '영화코드', how = 'left')
merge_left_code
```

```python
merge_left_code.to_csv("./movie_merge_code.csv", mode='w', encoding='utf-8', index=False) 
```

```python
merge_left_name.to_csv("./movie_merge_name.csv", mode='w', encoding='utf-8', index=False) 
```

```python
merge_left = pd.merge(list_name_rank, mv, on= '영화명(국문)', how = 'left')
merge_left
```

```python
genre_cnt = merge_genre.groupby(['장르'], as_index=False).sum().sort_values(by=['빈도수'], ascending=False)
genre_cnt = genre_cnt.reset_index(drop=True)
genre_cnt
```

```python
genre_scrn = merge_genre.groupby(['장르'], as_index=False).sum().sort_values(by=['상영 스크린 수'], ascending=False)
genre_scrn = genre_scrn.reset_index(drop=True)
genre_scrn
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x="장르", y="상영 스크린 수", data=genre_scrn)
plt.title("장르별 상영 스크린 수 분석")
plt.xticks(rotation=45)
plt.show()
```

## 배우

```python
merge_left_name
```

```python
actor_list = []
```

```python
for i in range(1,4) :
    for j in merge_left_name['배우{}'.format(i)]:
        actor_list.append(j)
```

```python
len(actor_list)
```

```python
actors = pd.DataFrame(actor_list,
             columns = ['배우'])
```

```python
actors['count'] = 1
actors
```

```python
actors_cnt = actors.groupby(['배우'], as_index=False).count().sort_values(by=['count'], ascending=False)
actors_cnt
```

```python
actors_cnt = actors_cnt.drop(actors_cnt[actors_cnt['배우'] == '없음'].index)
actors_cnt = actors_cnt.reset_index(drop=True)
actors_cnt
```

```python
actors_cnt_rank = actors_cnt.head(20)
actors_cnt_rank
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x="배우", y="count", data=actors_cnt_rank, color= 'gold')
plt.title("박스오피스 배우 빈도수 분석")
plt.xticks(rotation=45)
plt.show()
```
