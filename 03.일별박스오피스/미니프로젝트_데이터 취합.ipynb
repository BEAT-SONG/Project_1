## For문 돌려서 2017~2021 문화의날 

```python
c_list = []

for j in range(2017, 2022):
    for i in range(1,13):
        tmp = df_total_tmp[(df_total_tmp["년도"]==j) & (df_total_tmp["월"]==i) & (df_total_tmp["요일"]==2)]
        dept = tmp.loc[:,["년도","월","일자","관객수"]]
        tmp1 = dept.groupby(["년도","월","일자"]).sum().iloc[-1:].reset_index()
        c_list.append([tmp1['년도'].values[0],tmp1['월'].values[0],tmp1['일자'].values[0],tmp1['관객수'].values[0]])


Cultureday = pd.DataFrame(c_list, columns = ['년도', '월', '일자', '관객수'])


 
Cultureday
```

# 2017년 월별 문화의 날 제외한 나머지 수요일 평균

```python
#2017년 12월 수요일(2) 데이터 뽑기
Wednesday_2017_12 = df_total_tmp[(df_total_tmp["년도"]==2017) & (df_total_tmp["월"]==12) & (df_total_tmp["요일"]==2)]

#일자, 관객수만 추출
dept = Wednesday_2017_12.loc[:,["년도","월","일자","관객수"]]

#일자별 그룹화
tmp= dept.groupby(["년도","월","일자"]).sum()

#문화의 날 제외하고 나머지 평균 구하기
exclude_culture = tmp.iloc[:-1,].reset_index()

dd = exclude_culture.groupby(["년도","월"]).mean().reset_index()
exclude_culture_mean = dd.loc[:,["년도","월","관객수"]]


# 소숫점 처리
exclude_culture_mean["관객수"] = round(exclude_culture_mean["관객수"],0)
exclude_culture_mean
```

# 2017년 월별 문화의 날 제외한 나머지 수요일 평균

```python
c_list = []

for j in range(2017, 2022):
    for i in range(1,13):
        tmp = df_total_tmp[(df_total_tmp["년도"]==j) & 
                           (df_total_tmp["월"]==i) & 
                           (df_total_tmp["요일"]==2)]
        dept = tmp.loc[:,["년도","월","일자","관객수"]]
        
        # 일자별 그룹화
        tmp1 = dept.groupby(["년도","월","일자"]).sum()
        
        # 문화의날 제외하고 나머지 평균 구하기
        exclude_culture = tmp1.iloc[:-1,].reset_index()
        
        dd = exclude_culture.groupby(["년도","월"]).mean().reset_index()
        exclude_culture_mean = dd.loc[:,["년도","월","관객수"]]
       
        #소숫점 처리
        exclude_culture_mean["관객수"] = round(exclude_culture_mean["관객수"],0)
        
        
        c_list.append([exclude_culture_mean['년도'].values[0],
                       exclude_culture_mean['월'].values[0],
                       exclude_culture_mean['관객수'].values[0]])

ExcludeCultureday = pd.DataFrame(c_list, columns = ['년도', '월', '관객수'])
ExcludeCultureday
```
