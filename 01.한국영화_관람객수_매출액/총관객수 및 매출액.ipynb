```python
# https://www.kobis.or.kr/kobis/business/stat/them/findYearlyTotalList.do
# 총 관객 수 및 매출액(연도별)
```

```python
import pandas as pd
```

```python
file_path = "./KOBIS_총_관객수_및_매출액_연도별.xlsx"
df = pd.read_excel(file_path)
df
```

```python
df2 = df.drop(df.index[0:3])
df2
```

```python
df2.fillna('년도',inplace=True)
```

```python
df3 = df2.rename(columns=df2.iloc[0])
df3
```

```python
df3 = df3.drop(df3.index[0])
df3
```

```python
df4 = df3.query('년도 != "합계"')
df4
```

```python
df5 = df4.drop(22, axis=0)
df5
```

```python
df1 = df5.set_index("년도")
df1
```

```python
df2 = df1.iloc[:,[10,11,12,13]]
df2
```

```python
df3 = df2.tail(6)
df3
```

# 총 매출액 추이

```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
```

```python
df4 = df3[['매출액']]
```

```python
plt.rc('font', family = 'Malgun Gothic')
df4.plot.bar(title='국내 영화 연도별 매출액',
             rot = 0,
             figsize=(10,5),
            width = 0.5)
```

```python
df4.plot()
```

```python
df5 = df4.reset_index()

list(df5["매출액"])
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 3)
plt.rcParams['font.size'] = 12

plt.rc('font', family='Malgun Gothic', size = 10)

x = np.arange(2016, 2022)
y1 = np.array(list(df5["매출액"]))
y2 = np.array(list(df5["매출액"]))


fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0,2313989080068)
ax1.set_xlabel("년도")
ax1.set_ylabel("매출액")
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0,2313989080068)
#ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')

plt.show()
```
