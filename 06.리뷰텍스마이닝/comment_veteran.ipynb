```python
#페이지수 : 3950
coment_list = []

for j in range (1,3951):
    try:
        url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=115977&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={}".format(j)

        html = driver.page_source

        from bs4 import BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")



        for i in range (0,10):
            text = list(soup.select("#_filtered_ment_{}".format(i))[0])[0][64:]
            text_mod = text.replace("\t","")
            text_mod_final = text_mod.replace("\n","")
            
            coment_list.append(text_mod_final)
        
    except :
        pass
```

```python
#저장
import csv

with open('./movie_comment/veteran.txt', 'w', newline='') as f: 
    writer = csv.writer(f) 
    writer.writerow(coment_list)
```
