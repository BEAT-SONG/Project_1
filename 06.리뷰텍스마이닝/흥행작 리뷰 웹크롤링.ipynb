```python
# 웹 url로 접근하기
driver = webdriver.Chrome("C:/ChromeDriver/chromedriver.exe")
url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?=N&order=sympathyScore&page=1"
driver.get(url)
```

```python
coment_list = []
#페이지수 : 4726
for j in range (1,4727):
    try:
        url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?=N&order=sympathyScore&page={}".format(j)

        html = driver.page_source

        from bs4 import BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")



        for i in range (0,10):
            text = list(soup.select("#_filtered_ment_{}".format(i))[0])[0][64:]
            text_mod = text.replace("\t","")
            text_mod_final = text_mod.replace("\n","")
            
            coment_list.append(text_mod_final)
        
    except :
        pass
```

```python
coment_list
```

# .csv .txt로 저장

```python
import csv
with open('jobofhell.txt', 'w', newline='') as f: 
    writer = csv.writer(f) 
    writer.writerow(coment_list)

```
