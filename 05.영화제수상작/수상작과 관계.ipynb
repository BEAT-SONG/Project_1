# 수상작과 관람객 수 관계

```python
import pandas as pd
import numpy as np
import datetime
```

```python
file_path = "./total_movie_list_tmp.csv"
```

```python
df = pd.read_csv(file_path,encoding = "utf-8")
df
```

```python
df2 = df.sort_values('누적관객수',ascending=False)
df2
```

```python
a = df.drop_duplicates(["영화명(국문)"])
a
```

```python
a = a.reset_index()
a
```

```python
a
```

```python
a = a.sort_values('누적관객수',ascending=False)
```

```python
a["누적관객수순위"] = range(1,3892)
a
```

## 누적관객수순위 관계 (칸)

```python
b = a[['누적관객수순위','영화명(국문)','누적관객수']]
b
```

```python
q = b[b['영화명(국문)']== "기생충"]
```

```python
w = b[b['영화명(국문)']=="버닝"]
```

```python
e = b[b['영화명(국문)']=="아가씨"]
```

```python
r = b[b['영화명(국문)']=="시"]
```

```python
tmp = pd.concat([q,w,e,r], axis=0)
tmp
```

```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
```

```python
plt.rc('font', family='Malgun Gothic', size = 20)
plt.figure(figsize = (20,15))
plt.title("칸 누적 관객 수", size=30)
plt.xlabel("영화명", size=25)
plt.ylabel("누적관객수", size=25)

plt.plot(tmp["영화명(국문)"], 
         tmp["누적관객수"],
         color= "#FF6699", 
         linestyle='--',
         linewidth = 5,
         label = "내국인 영화 이용 카드금액")
```

```python
plt.rc('font', family='Malgun Gothic', size = 20)
plt.figure(figsize = (20,15))
plt.title("칸 누적 관객 수", size=30)
plt.xlabel("영화명", size=25)
plt.ylabel("누적관객수", size=25)

plt.bar(tmp["영화명(국문)"], 
         tmp["누적관객수"],
         color= "#FF6699", 
         label = "내국인 영화 이용 카드금액")
```
